```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# 
kB = 1.38e-23  
T1 = 300 
T_H = 500
T_L = 100
N = 6e23

# 
i = np.linspace(0,4e-21,1000)

beta_T1 = 1 / (kB * T1)
beta_T_H = 1 / (kB * T_H)
beta_T_L = 1 / (kB * T_L)

Z_T1 = np.sum(np.exp(-beta_T1 * i))
Z_T_H = np.sum(np.exp(-beta_T_H * i))
Z_T_L = np.sum(np.exp(-beta_T_L * i))

#  
Ni_T1 = (N / Z_T1) * np.exp(-beta_T1 * i)
Ni_T_H = (N / Z_T_H) * np.exp(-beta_T_H * i)
Ni_T_L = (N / Z_T_L) * np.exp(-beta_T_L * i)

# 
plt.plot(Ni_T1, i, label=f"T = {T1} K")
plt.plot(Ni_T_H, i, label=f"T = {T_H} K")
plt.plot(Ni_T_L, i, label=f"T = {T_L} K")
plt.xlabel("Number of Particles")
plt.ylabel("Energy Level i (J)")
plt.title("i vs Ni")
plt.grid()
plt.legend()
plt.xlim([0,2e21])

plt.savefig("C:\\Users\\user\\Downloads\\Boltzmann.jpg", format="jpg")
plt.show()
```
