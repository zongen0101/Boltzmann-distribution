```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# 常數
kB = 1.38e-23  
T1 = 300 
T_H = 500
T_L = 100
N = 6e23

# 範圍
εi = np.linspace(0,4e-21,1000)

beta_T1 = 1 / (kB * T1)
beta_T_H = 1 / (kB * T_H)
beta_T_L = 1 / (kB * T_L)

Z_T1 = np.sum(np.exp(-beta_T1 * εi))
Z_T_H = np.sum(np.exp(-beta_T_H * εi))
Z_T_L = np.sum(np.exp(-beta_T_L * εi))

# 粒子分佈 
Ni_T1 = (N / Z_T1) * np.exp(-beta_T1 * εi)
Ni_T_H = (N / Z_T_H) * np.exp(-beta_T_H * εi)
Ni_T_L = (N / Z_T_L) * np.exp(-beta_T_L * εi)

# 畫圖
plt.plot(Ni_T1, εi, label=f"T = {T1} K")
plt.plot(Ni_T_H, εi, label=f"T = {T_H} K")
plt.plot(Ni_T_L, εi, label=f"T = {T_L} K")
plt.xlabel("Number of Particles")
plt.ylabel("Energy Level εi (J)")
plt.title("εi vs Ni")
plt.grid()
plt.legend()
plt.xlim([0.5,0.75e21])

plt.savefig("C:\\Users\\user\\Downloads\\Boltzmann.jpg", format="jpg")
plt.show()
```
